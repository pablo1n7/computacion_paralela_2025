```python
import pandas as pd
from pathlib import Path

ped_path = Path("data_ch6/chr6_hg38_filter_prune_ped.ped")
ped_df = pd.read_csv(ped_path, delim_whitespace=True, header=None, dtype=str)

geno = ped_df.iloc[:, 6:].reset_index(drop=True) # Primeras 6 columnas: FID, IID, PAT, MAT, SEX, PHENO
alelo_uno = geno.iloc[:, 0::2]  # Columnas pares: alelo 1
alelo_dos = geno.iloc[:, 1::2]  # Columnas impares: alelo 2

#Exportar los alelos en dos archivos diferentes
alelo_uno.to_csv('data_ch6/chr6_allele1_wmeta.csv')
alelo_dos.to_csv('data_ch6/chr6_allele2_wmeta.csv')
```
